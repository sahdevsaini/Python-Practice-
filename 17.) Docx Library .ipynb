# Creating new doc document using docx module

```python
import docx as doc1                         # importing docx library as doc1
```

```python
document   = doc1.Document()                # Calling Method Document()  which is used creating word document and adding data into it.  
```

```python
document.add_paragraph("I'm a Aspiring Data Scientist")           # Adding a paragraph in to new create or exting word file.
document.add_paragraph("Hello World!") 
```

```python
document.save('Mynewdoc.docx')                                    # Saving a new word file with its name
```
